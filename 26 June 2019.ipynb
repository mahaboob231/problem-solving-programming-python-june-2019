```python
# Function to find the second largest number in a list
def secondLargest(li):
    # Convert the list into a unique list
    # O(n)
    unique = []
    for n in li:
        if n not in unique:
            unique.append(n)
    unique = sorted(unique,reverse=True)
    if len(unique)>1:
        return unique[1]
    else:
        return -1
secondLargest([1])

# def fifthleast

# def klargest

# def ksmallest


    
```

```python
def fifthleast(li):
    # Convert the list into a unique list
    # O(n)
    unique = []
    for n in li:
        if n not in unique:
            unique.append(n)
    unique = sorted(unique)
    return unique[4]
fifthleast([1,2,3,4,2,5,7,8,1,2])
```

```python
def kthLargest(li,k):
    # Convert the list into a unique list
    # O(n)
    unique = []
    for n in li:
        if n not in unique:
            unique.append(n)
    unique = sorted(unique,reverse=True)
    if k<=len(unique) and k!=0:
        return unique[k-1]
    else:
        return -1
kthLargest([1,2,3,4,5,6,7,8,9],0)
```

```python
# Function to identify the element with highest frequency in a list
#
# highestfrequency ([1,2,3,9,8,7,3,4,2,1])->1
# def highestFrequency(li):
#     unique = []
#     c = []
#     for n in li:
#         if n not in unique:
#             unique.append(n)
#             c.append(li.count(n))
#     d=[]
#     for i in range(0,len(unique)):
#         if c[i]==max(c):
#                    d.append(unique[i])
#     return min(d)
            
# highestFrequency([1,2,3,2,5,7,3,1,1,2,2])            
def highestFrequency(li):
    unique = {}
    for n in li:
        if n in unique:
            unique[n]+=1
        else:
            unique[n]=1
# Getting all frequencies into a list
    freq = unique.values()
    maxfreqE = []
    maxfreq = max(freq)
    # Identify the smallest number with maximum frequency
    for item in unique.items():
        if item[1]== maxfreq:
            maxfreqE.append(item[0])
    # Select the minimum from the keys
    return min(maxfreqE)
highestFrequency([1,2,3,2,5,7,3,1,1,2,2])

    
```
